```python
import pandas as pd
```

```python
data = pd.read_csv('spambase.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotting the line chart for the 'capitalLong' column
plt.figure(figsize=(10,6))
plt.plot(data['capitalLong'],marker='o', linestyle='-')
plt.title('Line Chart of capitalLong') 
plt.xlabel('Index') 
plt.ylabel('capitalLong') 
plt.grid(True) 
plt.show()
```

# Question 2: 

```python
cap = data['capitalLong'].value_counts()
plt.figure(figsize=(8, 6))
cap.plot(kind='bar') 
plt.title('Count of Each Type') 
plt.xlabel('Type')
plt.ylabel('Count') 
plt.xticks(rotation=0)
plt.show()
```
