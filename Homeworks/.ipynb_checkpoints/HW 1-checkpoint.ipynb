Question 1: Create a line chart to visualize the distribution of the 'capitalLong' feature.

```python
import pandas as pd
```

```python
data = pd.read_csv('spambase.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotting the line chart for the 'capitalLong' column
plt.figure(figsize=(10,6))
plt.plot(data['capitalLong'],marker='o', linestyle='-')
plt.title('Line Chart of capitalLong') 
plt.xlabel('Index') 
plt.ylabel('capitalLong') 
plt.grid(True) 
plt.show()
```

Question 2: Generate a bar chart to show the frequency of 'type' (spam and nonspam) in the dataset. 

```python
cap = data['type'].value_counts()
plt.figure(figsize=(8, 6))
cap.plot(kind='bar') 
plt.title('Count of Each Type') 
plt.xlabel('Type')
plt.ylabel('Count') 
plt.xticks(rotation=0)
plt.show()
```

Question 3: Create a scatter plot to explore the relationship between 'num3d' and 'num000'

```python
plt.figure(figsize=(8, 6))  # Set figure size
plt.scatter(data['num3d'], data['num000'], alpha=0.6)  # Plot with some transparency (alpha)
plt.title('Scatter Plot of num3d vs num000')
plt.xlabel('num3d')
plt.ylabel('num000')
plt.grid(True)  # Optionally add gridlines
plt.show()
```

Question 4: Visualize the distribution of 'money' using a histogram.

```python
# Sample data for 'money'
money = [100, 200, 150, 120, 180, 220, 250, 300, 400, 450, 500, 550, 600]
# Creating a histogram
plt.hist(money, bins=10, edgecolor='black')  # 'bins' controls the number of bars in the histogram
plt.title('Distribution of Money')
plt.xlabel('Amount of Money')
plt.ylabel('Frequency')
plt.show()
```

Question 5: Create a pie chart to represent the proportion of spam and nonspam emails in the dataset.

```python
spam_count = data['type'].value_counts()[1]
nonspam_count = data['type'].value_counts()[0]

# Data for the pie chart
labels = ['Spam', 'Non-Spam']
sizes = [spam_count, nonspam_count]

# Creating the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Spam vs Non-Spam Emails')
plt.show()
```

Question 6:  Generate a table chart to display the summary statistics for the 'capitalAve' feature.

```python
# Calculate summary statistics for 'capitalAve'
summary_stats = data['capitalAve'].describe()

# Create a figure and axis
fig, ax = plt.subplots()

# Hide the axes for a clean table
ax.axis('off')

# Display the summary statistics in a table
table = ax.table(cellText=summary_stats.values.reshape(-1, 1),
                 rowLabels=summary_stats.index,
                 colLabels=['capitalAve'],
                 cellLoc='center',
                 loc='center')

# Set the font size for better readability
table.auto_set_font_size(False)
table.set_fontsize(12)

# Adjust column width
table.scale(1.5, 1.5)

plt.title("Summary Statistics for 'capitalAve'")
plt.show()
```

Question 7: Calculate the mean, median, and standard deviation of the 'our' feature.

```python
# Calculate mean, median, and standard deviation for the 'our' feature
mean_our = data['our'].mean()
median_our = data['our'].median()
std_dev_our = data['our'].std()

# Display the results
print(f"Mean of 'our': {mean_our}")
print(f"Median of 'our': {median_our}")
print(f"Standard Deviation of 'our': {std_dev_our}")
```

Question 8: Determine the minimum and maximum values of 'hp'.

```python
# Determine the minimum and maximum values for the 'hp' feature
min_hp = data['hp'].min()
max_hp = data['hp'].max()

# Display the results
print(f"Minimum value of 'hp': {min_hp}")
print(f"Maximum value of 'hp': {max_hp}")
```

Question 9: Calculate the 25th, 50th, and 75th percentiles for 'telnet'.

```python
# Calculate the 25th, 50th, and 75th percentiles for the 'telnet' feature
percentiles = data['telnet'].quantile([0.25, 0.5, 0.75])

# Display the results
print("25th percentile of 'telnet':", percentiles[0.25])
print("50th percentile (median) of 'telnet':", percentiles[0.5])
print("75th percentile of 'telnet':", percentiles[0.75])
```

Question 10: Find the mode of the 'you' feature.

```python
# Find the mode of the 'you' feature
mode_you = data['you'].mode()

# Display the result
print("Mode of 'you':", mode_you.values)
```

Question 11:calculate the interquartile range (IQR) for 'business'.

```python
# Calculate the 25th and 75th percentiles (Q1 and Q3)
Q1 = data['business'].quantile(0.25)
Q3 = data['business'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Display the result
print("Interquartile Range (IQR) for 'business':", IQR)
```

Question 12: Compute the skewness and kurtosis of the 'credit' feature.

```python
from scipy.stats import skew, kurtosis
```

```python
# Calculate skewness and kurtosis for the 'credit' feature
skewness = skew(data['credit'])
kurtosis_value = kurtosis(data['credit'])

# Display the results
print("Skewness of 'credit':", skewness)
print("Kurtosis of 'credit':", kurtosis_value)
```

Question 13: Calculate the correlation coefficient between 'capitalAve'and'capitalTotal'.

```python
# Calculate the correlation coefficient between 'capitalAve' and 'capitalTotal'
correlation_coefficient = data['capitalAve'].corr(data['capitalTotal'])

# Display the result
print("Correlation coefficient between 'capitalAve' and 'capitalTotal':", correlation_coefficient)
```
